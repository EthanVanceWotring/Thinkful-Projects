```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
pd.set_option('display.max_columns', 500)
```

```python
president_primary_polls = pd.read_csv('president_primary_polls.csv')
```

```python
president_primary_polls.head()
```

```python
president_primary_polls.info()
```

```python
president_primary_polls['pollster_rating_name'].value_counts()
```

```python
president_primary_polls_YouGov = president_primary_polls[president_primary_polls['pollster_rating_name'] == 'YouGov']
```

```python
president_primary_polls_YouGov.head()
```

##### Question 1

Choose one variable and plot that variable four different ways.

    * I am going to plot the variable 'candidate_name' four different ways

```python
# Setting the overall aesthetic.
sns.set(style="darkgrid")

# Use catplot (previous versions of seaborn use factorplot)
g = sns.catplot(x="candidate_name", y="pct", hue="pollster_rating_name", data=president_primary_polls,
                   height=100, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Rating")
g.set_xlabels("")
plt.title('All Polls in Dataset')
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="darkgrid")

# Use catplot (previous versions of seaborn use factorplot)
g = sns.catplot(x="candidate_name", y="pct", hue="pollster_rating_name", data=president_primary_polls_YouGov,
                   height=100, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Rating")
g.set_xlabels("")
plt.title('YouGov Poll')
plt.show()
```

```python
president_primary_polls_YouGov_11_19 = president_primary_polls_YouGov[president_primary_polls_YouGov['end_date'] == '11/26/19']
```

```python
# Setting the overall aesthetic.
sns.set(style="darkgrid")

# Use catplot (previous versions of seaborn use factorplot)
g = sns.catplot(x="candidate_name", y="pct", hue="pollster_rating_name", data=president_primary_polls_YouGov_11_19,
                   height=100, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Rating")
g.set_xlabels("")
plt.title('YouGov Poll 11/26/19')
plt.show()
```

##### Scatter Plot

```python
sns.scatterplot(x='answer',y='pct',data=president_primary_polls_WarrenBiden)
```

##### Line Plot

```python
g = sns.relplot(x="answer", y="pct", kind="line", data=president_primary_polls_WarrenBiden)
g.fig.autofmt_xdate()
```

```python
sns.tsplot(data=president_primary_polls_WarrenBiden, time="end_date",
           condition="candidate_name", value="pct")
plt.show()


#sns.tsplot(df_open, time='Date', unit='Dummy', condition='Company', value='Price')
```
